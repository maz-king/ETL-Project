```python
# Code Source
# https://stackoverflow.com/questions/52688645/valueerror-time-data-10-11-2006-2400-does-not-match-format-d-m-y-hm

df['datetime_zero'] = df['datetime'].str.replace('24:00', '0:00')
df['datetime_er'] = pd.to_datetime(df['datetime_zero'], format='%m/%d/%Y %H:%M')
selrow = df['datetime'].str.contains('24:00')
df['datetime_obj'] = df['datetime_er'] + selrow * timedelta(days=1)
```

```python
dates=[]
for index, row in df.iterrows():
    date=row['datetime_obj']
    #date_df[index] = datetime.strptime(df['datetime'][index], "%m/%d/%Y %H:%M")
    #strip_date=datetime.strptime(dates, '%m/%d/%Y %H:%M').date()
    #strip_date=dates.date()
    strip_date_object=datetime.strftime(date, '%m/%d/%Y')
    #strip_date_object.date()
    dates.append(strip_date_object)
```

```python
df['date']=dates
```
